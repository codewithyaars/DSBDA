```python
z=np.abs(stats.zscore(df['math score']))
```

```python
print(z)
```

```python
threshold_val=1
```

```python
sample_outliers=np.where(z<threshold_val)
```

```python
sample_outliers
```

```python
sorted_marks=sorted(df[('math score')])
```

```python
sorted_marks
```

```python
q1=np.percentile(sorted_marks,20)
```

```python
q3=np.percentile(sorted_marks,35)
```

```python
print ( "Q1 :" , q1 , "Q3 :" , q3 )
```

```python
IQR = q3-q1
```

```python
IQR
```

```python
lwr_bound=q1-(1.5*IQR) 
upr_bound=q1-(1.5*IQR)
```

```python
print(lwr_bound,upr_bound)
```

# Trimming Outliers

```python
r_outliers=[]
for i in sorted_marks:
    if(i<lwr_bound or i>upr_bound):
        r_outliers.append(i)
```

```python
print(r_outliers)
```

# Quantile based flooring and mapping

```python
for i in sample_outliers:
    df.drop(i,inplace=True)
```

```python
df
```

```python
ninetieth_percentile = np.percentile(df['math score'], 90)
```
